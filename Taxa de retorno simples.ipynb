```python
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
start_date = '2015-01-01'    # format: aaaa/mm/dd/
end_date = ''                # format: aaaa/mm/dd/
EMBR3 = wb.DataReader('EMBR3.SA', data_source='yahoo', start=start_date)
EMBR3.head()
```

# Calculando retorno de uma ação

$\mathbf{\text{Taxa de retorno simples}}$
$$Taxa_{simples} = \frac{Preço_{final}-Preço_{inicial}}{Preço_{inicial}} = \left(\frac{Preço_{final}}{Preço_{inicial}} - 1\right)*100\%$$

```python
EMBR3['Retorno simples'] = 100*((EMBR3['Adj Close']/EMBR3['Adj Close'].shift(1)) - 1)
# .shift(n) é função do pandas para uso em dataframes

print(EMBR3['Retorno simples'])
# a primeira célula tem valor NaN por não ter registro de um dia anterior ao primeiro dia do DataFrame
```

```python
EMBR3['Retorno simples'].plot(figsize=(16,10))
plt.show()
```
